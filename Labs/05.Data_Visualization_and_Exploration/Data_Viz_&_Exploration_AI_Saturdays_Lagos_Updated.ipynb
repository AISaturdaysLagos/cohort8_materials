### Removing special characters

<p>By looking at a random sample of the dataset rows, we observe that some entries in the columns like <code>Installs</code> and <code>Price</code> have a few special characters (<code>+</code> <code>,</code> <code>$</code>) due to the way the numbers have been represented. This prevents the columns from being purely numeric, making it difficult to use them in subsequent future mathematical calculations. Ideally, as their names suggest, we would want these columns to contain only digits from [0-9].</p>
<p>Hence, we now proceed to clean our data. Specifically, the special characters <code>,</code> and <code>+</code> present in <code>Installs</code> column and <code>$</code> present in <code>Price</code> column need to be removed.</p>

```python
# List of characters to remove
chars_to_remove = ['+',',','$']
# List of column names to clean
cols_to_clean = ['Installs', 'Price']

# Loop for each column in cols_to_clean
for col in cols_to_clean:
    # Loop for each char in chars_to_remove
    for char in chars_to_remove:
        # Replace the character with an empty string
        apps_data[col] = apps_data[col].apply(lambda x: x.replace(char, ''))
```

```python
# You can also use regular expression
# apps_data['Price'] = apps_data['Price'].str.replace('[$]', '', regex=True)
```

### Correcting data types

```python
# Convert Installs to float data type
apps_data['Installs'] = apps_data['Installs'].astype(int)

# Convert Price to float data type
apps_data['Price'] = apps_data['Price'].astype(float)

# Checking dtypes of the apps dataframe
apps_data.dtypes
```

```python
# Conver the Last Updated feature to a datetime datatype
apps_data['Last Updated'] = pd.to_datetime(apps_data['Last Updated'], format='%B %d, %Y')
```

### Creating a wrangle function

```python
def wrangle_data(data_path):
    """
    Data Loading and Cleaning:
    - Load the dataset from the specified data file.
    - Perform data cleaning steps such as handling missing values, removing duplicates, and fixing data types.

    Parameters:
        data_path (str): The file path to the dataset.

    Returns:
        pd.DataFrame: Cleaned and processed DataFrame.
    """
    apps_data = pd.read_csv(data_path)
    apps_data.drop(columns=["Unnamed: 0"], inplace=True)

    # List of characters to remove
    chars_to_remove = ['+',',',"$"]
    # List of column names to clean
    cols_to_clean = ['Installs', 'Price']

    # Loop for each column in cols_to_clean
    for col in cols_to_clean:
        # Loop for each char in chars_to_remove
        for char in chars_to_remove:
            # Replace the character with an empty string
            apps_data[col] = apps_data[col].apply(lambda x: x.replace(char, ''))

    # Convert Installs to float data type
    apps_data['Installs'] = apps_data['Installs'].astype(int)

    # Convert Price to float data type
    apps_data['Price'] = apps_data['Price'].astype(float)

    # Convert `Last Updated`` to datetime data type
    apps_data['Last Updated'] = pd.to_datetime(apps_data['Last Updated'], format='%B %d, %Y')


    return apps_data
```

```python
# Call the wrange_data function
apps_data = wrangle_data("data/apps.csv")
apps_data.tail()
```

## Data Exploration and Visualization

1. Which category has the highest share of (active) apps in the market?


```python
num_categories = apps_data['Category'].unique()
print('Number of categories = ', num_categories)

# Count the number of apps in each 'Category'.
num_apps_in_category = apps_data.Category.value_counts()
print(num_apps_in_category)
```

We see that the `FAMILY` Category has the highest number of apps while the `BEAUTY` Category has the lowest number of apps on Google Play Store

```python
plt.figure(figsize=(15, 6)) 
num_apps_in_category.plot(x = num_apps_in_category.index, y = num_apps_in_category.values, kind='bar');
```

Pandas does basic visualizations but we have the liberty to customize our the visuals when we use solid visulization libraries like seaborn

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 6)

ax = sns.barplot(x = num_apps_in_category.index, y = num_apps_in_category.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# Add x and y label axis
plt.xlabel("Catergory")
plt.ylabel("Number of Apps")

# Add plot title
plt.title("Barplot showing the Share of apps per Category");
```

The seaborn barplot above is better than that of pandas but plotly is best for visualizing categorical variables

```python
# Using plotly
fig = px.bar(
        x = num_apps_in_category.index, # index = category name
        y = num_apps_in_category.values, # value = count
)

fig.update_layout(
    title='Barplot showing the Share of apps per Category',
    xaxis_title='Categories',
    yaxis_title='Count'
)

fig.show()
```

2. Which app categories are the most popular in terms of the number of installations?


```python
# Group by Category and sum the installs
category_installs = apps_data.groupby('Category')['Installs'].sum().sort_values(ascending=False).reset_index()

# Create a horizontal bar chart
fig = px.bar(category_installs, x="Installs", y="Category", color="Category",
             title='Most Popular App Categories by Installs')
fig.show()
```

We can then see here that although the `GAME` Category did not have the highest apps on Google Play Store, it has the highest number of installations across all categories

Let's check for the general correlation between the features

Heatmaps are often used to visualize correlation matrices. In this context, they can show the strength and direction of relationships between variables. High positive correlations are represented by one color, while high negative correlations are represented by another.

```python
apps_data.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(apps_data.corr(numeric_only=True), annot=True)
plt.title("Heatmap");
```

 The correlation score displayed in each cell represents the strength and direction of the linear relationship between the two variables. For instance, the correlation score of `Installs` and `Reviews` is `0.63` and this indicates a `positive` correlation. A positive correlation score close to 1 indicates a strong positive linear relationship between the two variables. When one variable increases, the other tends to increase as well, and vice versa. A negative correlation score close to -1 indicates a strong negative linear relationship between the two variables. When one variable increases, the other tends to decrease, and vice versa. A correlation score close to 0 indicates a weak or no linear relationship between the two variables. Changes in one variable do not significantly impact the other, and there is no clear linear pattern.

3. Is there a relationship between app price and the number of installations, and how does it differ by content rating?

```python
apps_without_outliers = apps_data.query("Installs<=1_000_000")
```

```python
# Create scatterplots of Price vs. Installs by Content Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Installs', hue='Content Rating', data=apps_without_outliers, palette='Set2')

# Add label and axis titles
plt.xlabel('Price')
plt.ylabel('Installs')
plt.title('Price vs. Installs by Content Rating')

# Add legend
plt.legend(title='Content Rating')

# show plot
plt.show()
```

There is no significant linear relationship between the price of an app and the number of installations as seen in the Heatmep and again in this scatter plot. But, it is neccary to note that there are more free apps on Google play store, about `90%` than Paid and this class imbalance could influence the correlation score hence we cannot arrive at a conclusion from this scatterplot


4. How have user reviews/installs for apps changed over time?

```python
# Group by year and plot the number of user installs over time
installs_over_time = apps_data.groupby(apps_data['Last Updated'].dt.year)['Installs'].sum()

plt.figure(figsize=(10,6))
plt.plot(installs_over_time.index, installs_over_time.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Installs')
plt.title('User Installs Over Time')
plt.show()
```

From 2010 to 2017, Apps installations on Google play store staedily inccreased until 2018 where there was a very sharp increase from about 3 biliion to about 70 billion installations. The reason behind this sharp increase can be further investigated.

5. Is there a correlation between app price and ratings, and does it vary by app category?

    Pairplot creates a grid of scatterplots, histograms, or other plots to visualize the pairwise relationships between the variables.


```python
# Create a scatterplot matrix
img = sns.pairplot(apps_data, vars=['Price', 'Rating'], hue='Category', diag_kind='kde')

# Adjust the subplot spacing
img.fig.subplots_adjust(top=0.95, hspace=0.3, wspace=0.3)  # Increase the 'top' value to adjust the super title position

# Add the super title
img.fig.suptitle('Price vs. Ratings by Category',y=1.02, x=0.5, ha='right')

# Show plot
plt.show()
```

A pairplot can also be used to visualize trend or relationship beteween variable. On the diagonal axis, a kde plot is used to visulize the distribution of the variables. A Kernel Density Estimation (KDE) plot is a data visualization technique used to estimate and visualize the probability density function of a continuous random variable. It helps to visualize the distribution of data. Here, we attempt to visualize the relationship between the Price and Rating of the apps on Google Play Store. Again, it appears that there are no linear relationship between both features but it was observed that paid apps are not rated below 3 on the rating scale.

- Distribution of app ratings

```python
# Average rating of apps
avg_app_rating = apps_data['Rating'].mean()
print('Average app rating = ', avg_app_rating)

app_rating_median = apps_data["Rating"].median()
print('Median of the app rating = ', app_rating_median)

# Create a histogram to show the distribution of app ratings
fig = px.histogram(apps_data, x='Rating', title='App Rating Distribution') 

# Add a vertical dashed line to indicate the average app rating
fig.add_shape(
    type='line',
    x0=avg_app_rating,
    y0=0,
    x1=avg_app_rating,
    y1=1000,
    line=dict(dash='dashdot')
)

# Show the plot
fig.show()
```

We can see here that the distribution of app rating on Google play store is negatively skewed(Left Skew) as there are more apps with high ratings and the median: `4.3` is greater than the mean: `4.17`.

6. How does the sentiment of user reviews vary across different app categories, and which categories tend to have more positive reviews?

```python
# Load user_reviews.csv
reviews_df = pd.read_csv("data/user_reviews.csv")

# Join the two dataframes
merged_df = pd.merge(apps_data, reviews_df, on='App')
```

```python
reviews_df.head()
```

**data_reviews.csv**

| variable               | class        | description                                           |
|:-----------------------|:-------------|:------------------------------------------------------|
| App                    | character    | The application name                                  |
| Translated_Review      | character    | User review (translated to English)                   |
| Sentiment              | character    | The sentiment of the user - Positive/Negative/Neutral |
| Sentiment_Polarity     | character    | The sentiment polarity score                          |
| Sentiment_Subjectivity | character    | The sentiment subjectivity score                      |

```python
merged_df.head()
```

```python
# Calculate average sentiment polarity by category
avg_sentiment_by_category = merged_df.groupby('Category')['Sentiment_Polarity'].mean().reset_index()

# Create an interactive bar chart
fig = px.bar(avg_sentiment_by_category, x='Category', y='Sentiment_Polarity',
             title='Average Sentiment Polarity by App Category',
             labels={'Sentiment_Polarity': 'Average Sentiment Polarity'})
fig.show()
```

Although the `GAME` Category has the most installation, hthe Category happens to have the lowest average sentiment polarity, this means that apps in this category get the most nagative reviews based on the sentiment polarity score.
