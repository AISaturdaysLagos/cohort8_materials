## **Designing a Basic Neural Network to Predict Flower Models**

```python
import pandas as pd
from sklearn.datasets import load_iris


iris_data = load_iris()
features = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
features.head(15)
```

```python
iris_data.target
```

```python
#feature engineering: normalization
from sklearn.preprocessing import normalize
features = normalize(features)
features
```

```python
#data split:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, iris_data.target, random_state=42)
```

```python
#One-hot encoding the features
from keras.utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
y_train_cat
```

### **Model Development**

```python
#model development
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=64, input_dim=4, activation='relu'))         # Input Layer
model.add(Dense(units=64, activation='relu'))                      # First Hidden Layer
model.add(Dense(units=3, activation='softmax'))                   # Output Layer
```

Activation functions:
Relu - introduces non-linearity, and used to eliminate the vanishing gradient problem
Softmax - used for multi-class problems
Sigmoid - Used for binary classification problems

```python
model.summary()
```

Number of Parameters=(Number of Input Features+1)×Number of Neurons in the Layer

> Layer 1: (4 + 1) x 64 = 320

> Layer 2: (64 + 1) × 64= 4160

> Layer 3: (64 + 1) × 3 = 195

```python
#compile the model
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

**Optimizer options**: adam, adagrad, adadelta, nadam, adamax, SGD

**Loss options**: categorical_crossentropy - for multi-class problems
binary_crossentropy - for two-class problems
mean_squared_error, mean absolute error - for regression problems

**Metrics**: accuracy, precision, F1 score, recall, confusion_matrix



```python
#fit the model
history = model.fit(X_train, y_train_cat, batch_size = 10, epochs = 50, verbose = 1, validation_data= (X_test, y_test_cat))
```

**batch_size**
>
**Epochs**
> The number of times the training set is passed back and forth in the neural network, each epoch consists of a forward pass and a backward pass of the training set.

**Verbose:**
> The extent of details displayed during training; 0,1,2






Training accuracy
Validation accuracy


```python
# plot the history of the training
import matplotlib.pyplot as plt

# summarize the history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()

# summarize the history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
#print the shape of the train image set
print(train_images.shape)

#print the shape of the test image set
print(test_images.shape)

#print the shape of the train label set
print(train_labels.shape)

#print the shape of the test label set
print(test_labels.shape)

#print the length of the train image set
print(len(train_labels))

#print the length of the test image set
print(len(test_labels))
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Map each label to its corresponding class name
class_labels = [class_names[label] for label in train_labels]

# Create a Pandas DataFrame with two columns
labeled = pd.DataFrame({"count": train_labels, "class": class_labels})

# Group the DataFrame by class and count the number of occurrences
class_counts = labeled.groupby("class").count()
class_counts
```

```python
# Create a Plotly figure showing the first image in the train_images set
import plotly.express as px
fig = px.imshow(train_images[0], title="First image in the train_images set")

# Add a colorbar to the figure
fig.update_layout(coloraxis_colorbar=dict(title="Color Scale"), template= 'plotly_dark')

# Show the figure
fig.show()
```

```python
# one hot encode test and train target values
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
#Review the new label shapes
print(train_labels.shape)
print(test_labels.shape)
```

```python
#Normalizing the data
#Dividing by 255 which is the maximum number. This is to keep the pixel ranges between zero and one.
train_images = train_images / 255.0

test_images = test_images / 255.0
```

# **Model Training**

```python
from keras.layers import Flatten
from keras.layers import BatchNormalization
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))          #Input shape of 28 by 28
model.add(BatchNormalization())                   #Add a normalization layer
model.add(Dense(150, activation='relu'))         #Add a dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dropout(0.2))                        #Add a dropout layer
model.add(Dense(150, activation='relu'))        #Another dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dropout(0.2))                           #Add a dropout layer
model.add(Dense(150, activation='relu'))        #Another dense layer
model.add(BatchNormalization())                  #Another normalization layer
model.add(Dense(10, activation='softmax'))         #Output layer with 10 units
```

```python
#Complie the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


#Fit the model
history = model.fit(
    train_images,    # input training data
    train_labels,    # output training data
    batch_size=50,   # mini-batch gradient descent size
    epochs=15,       # number of iterations over the entire training data
    verbose=2,        # what type of information is printed during training
    validation_data=(test_images, test_labels))  # input test data, output test data
```

```python
model.summary()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create subplots for accuracy and loss
fig = make_subplots(rows=1, cols=2, subplot_titles=("Model Accuracy", "Model Loss"))

# Add accuracy plots to subplots
fig.add_trace(
    go.Scatter(x=list(range(1,len(history.history['accuracy'])+1)), y=history.history['accuracy'], mode='lines', name='Train Accuracy'),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=list(range(1,len(history.history['val_accuracy'])+1)), y=history.history['val_accuracy'], mode='lines', name='Test Accuracy'),
    row=1, col=1
)

# Add loss plots to subplots
fig.add_trace(
    go.Scatter(x=list(range(1,len(history.history['loss'])+1)), y=history.history['loss'], mode='lines', name='Train Loss'),
    row=1, col=2
)
fig.add_trace(
    go.Scatter(x=list(range(1,len(history.history['val_loss'])+1)), y=history.history['val_loss'], mode='lines', name='Test Loss'),
    row=1, col=2
)

# Update subplot properties
fig.update_xaxes(title_text='Epoch', row=1, col=1, range=[-1, len(history.history['accuracy'])-1])
fig.update_xaxes(title_text='Epoch', row=1, col=2, range=[-1, len(history.history['loss'])-1])
fig.update_yaxes(title_text='Accuracy', range=[0,1], row=1, col=1)
fig.update_yaxes(title_text='Loss', row=1, col=2)



# Update figure layout and display
fig.update_layout(title_text="Model Performance", height=500, width=1000, template= 'plotly_dark')
fig.show()
```
